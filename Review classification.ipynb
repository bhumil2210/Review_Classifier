```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer


def preprocessing_text(text_to_process):
    list_to_return=[]
    for i in range(0,len(text_to_process)):
        # column : "Review", row ith 
        #print(i)
        if not text_to_process[i]:
            del text_to_process[i]
        #print(fingerprint_set[i][0][0])
        review = re.sub('[^a-zA-Z]', ' ', str(text_to_process[i][0]))  

        # convert all cases to lower cases 
        review = review.lower()  

        # split to array(default delimiter is " ") 
        review = review.split()  

        # creating PorterStemmer object to 
        # take main stem of each word 
        ps = PorterStemmer()  

        # loop for stemming each word 
        # in string array at ith row     
        review = [ps.stem(word) for word in review 
                    if not word in set(stopwords.words('english'))]  

        # rejoin all string array elements 
        # to create back into a string 
        review = ' '.join(review)   

        # append each string to create 
        # array of clean text  
        list_to_return.append(review)
    return list_to_return
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
processted_master_list = preprocessing_text(shuffled_master_list)
cv_new=CountVectorizer()
X_new=cv_new.fit_transform(processted_master_list).toarray()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

def classification_model(operation_perform,list_to_train):
    cvi=CountVectorizer()
    X_i=cvi.fit_transform(operation_perform).toarray()
    y_new_i=[]
    for i in range(0,len(list_to_train)):
        y_new_i.append(list_to_train[i][1])
    print(len(X_i))
    print(len(y_new_i))
    X_train, X_test, y_train, y_test = train_test_split(X_i,np.asarray(y_new_i),test_size=0.2)
    model=RandomForestClassifier(n_estimators=501, criterion='entropy')
    model.fit(X_train,y_train)
    
    model_pred=model.predict(X_test)
    cm=confusion_matrix(y_test,model_pred)
    print(cm)
    print(accuracy_score(model_pred,y_test))
```

```python
corpus_battery = preprocessing_text(battery_life_sentiment_list)
battery_model = classification_model(corpus_battery,battery_life_sentiment_list)

corpus_picture_quality = preprocessing_text(picture_quality_sentiment_list)
picture_quality_model = classification_model(corpus_picture_quality,picture_quality_sentiment_list)

corpus_money = preprocessing_text(money_sentiment_list)
money_model = classification_model(corpus_money,money_sentiment_list)

corpus_sound_quality = preprocessing_text(sound_quality_sentiment_list)
sound_quality_model = classification_model(corpus_sound_quality,sound_quality_sentiment_list)

corpus_fingerprint = preprocessing_text(fingerprint_sentiment_list)
fingerprint_model = classification_model(corpus_fingerprint,fingerprint_sentiment_list)
```
